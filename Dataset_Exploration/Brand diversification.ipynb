```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('EMPORIOSIRENUSE_20173012/sales_2016-01-01_2017-12-21.csv',sep=",")
```

```python
data
```

```python
a = data.groupby('Billing city').count()['gross_sales']
adf = a.sort_values().reset_index()
adf = adf.sort_values(by = ['gross_sales'], ascending=False)
adf.to_csv('cities_sales_count.csv')
adf
```

```python
a = data.groupby('Billing city').sum()['gross_sales']
adf = a.sort_values().reset_index()
adf = adf.sort_values(by = ['gross_sales'], ascending=False)
adf.to_csv('cities_gross_sales.csv')
adf
```
