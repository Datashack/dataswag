```python
df[['likes_count','followers']] = df[['likes_count','followers']].apply(pd.to_numeric)

df['likes_score'] = (df['likes_count'])/(df['followers'])
```

```python
def zscore(x,mu,std):
    zscore = (x-mu)/std
    return(zscore)
```

```python
df['zscore'] = zscore(df['likes_score'], df.likes_score.std(), df.likes_score.mean())
```
